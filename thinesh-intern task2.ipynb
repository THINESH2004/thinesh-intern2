```python
Empd['Attrition']= Empd['Attrition'].replace({'No':0, 'Yes':1})
```

```python
Empd['OverTime']= Empd['Attrition'].map({'No':0, 'Yes':1})
Empd['Gender']= Empd['Gender'].map({'No':0, 'Yes':1})
```

```python
from sklearn.preprocessing import LabelEncoder

encoding_cols = ['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus']

for column in encoding_cols:
    label_encoder = LabelEncoder()
    Empd[column] = label_encoder.fit_transform(Empd[column])
```
